```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inacc = [6,30,10,50,60,50,70,80]
inloss = [89,75,63,45,39,24,21,19,13,9]
epochs = 10

eplabels = []
losslabels = []
```

```python
for i in range(0, len(inacc)):
    eplabels.append(i*epochs)
for i in range(0, len(inloss)):
    losslabels.append(i*epochs)
print(inacc, losslabels)
```

```python
plt.plot(eplabels,inacc, 'b')
plt.plot(losslabels,inloss, 'r')
plt.ylabel('Accuracy in %')
plt.xlabel('No. of Epochs')
plt.show()
```
