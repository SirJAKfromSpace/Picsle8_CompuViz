```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import sklearn.cluster as sk
from PIL import Image
from torchvision import transforms

import os
import glob
import os.path as osp
```

```python
def displayimage(imgarr, r=1, c=3):
    plt.figure(figsize=(10,10))
    count = 1
    for i in range(1,r+1):
        for j in range(1,c+1):
            plt.subplot(r,c,count)
            plt.imshow(imgarr[count-1])
            count += 1
    plt.show() 
```

```python
img = cv2.cvtColor(cv2.imread('images/nur.jpg'), cv2.COLOR_BGR2RGB)
downed = cv2.resize(img, dsize=(300, 300), interpolation=cv2.INTER_CUBIC)
edgesdowned = cv2.Canny(downed,200,200)

imgp = cv2.cvtColor(cv2.imread('images/pixb1.jpg'), cv2.COLOR_BGR2RGB)
downedp = cv2.resize(imgp, dsize=(300, 300), interpolation=cv2.INTER_CUBIC)
edgesdownedp = cv2.Canny(downedp,400,400)
laplacian = cv2.Laplacian(cv2.imread('images/pixb1.jpg',0),cv2.CV_64F)

plt.figure(figsize=(10,10))

plt.subplot(221),plt.imshow(img,cmap = 'gray')
plt.title('Original Real'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(edgesdowned,cmap = 'gray')
plt.title('Edge Real'), plt.xticks([]), plt.yticks([])

plt.subplot(223),plt.imshow(imgp,cmap = 'gray')
plt.title('Original Pix'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(edgesdownedp,cmap = 'gray')
plt.title('Edge Pix'), plt.xticks([]), plt.yticks([])

plt.show()
```
