```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inacc = [6,30,10,50,60,50,70,80]
epochs = 10

eplabels = []
for i in range(0, len(inacc)):
    eplabels.append(i*epochs)
print(inacc, eplabels)
```

```python
plt.plot(eplabels,inacc)
plt.ylabel('Accuracy in %')
plt.xlabel('No. of Epochs')
plt.show()
```
